```python
#check if given regex is present in the text
def matches_regex(regex, text):
    return bool(re.compile(regex).search(text))

#check if the text contains code
def contains_code(text):
    # filter based on keywords that indicate code
    code_blacklist = ['&&', '||', '<html>', 'SELECT']
    
    return (
            any(code_keyword in text for code_keyword in code_blacklist) |
            matches_regex(r'\w+\(\w*\) \{', text) | # e.g. myFunc() {
            matches_regex(r'def \w+\(', text) | # e.g. def parse_list(
            matches_regex(r'\[A-z]+\.[A-z]+', text) | # e.g. this.language
            matches_regex(r': [\w\.#]{1,12};', text) | # e.g. font-size: 1.3em;
            matches_regex(r'<\/\w+>', text)  # e.g. </html>
    )

#check if the text contains words
def contains_words(text):
    return matches_regex(r'[A-z]{3,}', text) # words with at least three characters

```

```python
def translate_text(text):

    return GoogleTranslator(source='auto', target='mt').translate(text)

    
def check_text(text):
    #check if the text contains code
    if contains_code(text):
        return False
    
    #check if the text contains words
    if not contains_words(text):
        return False

    return True
```

```python
f = open(r'data\alpaca_gpt4_data_mt.json', 'a')

mt_data = []

for i in range(49438, len(en_data)):

    #add 5 second delay to avoid getting blocked
    time.sleep(6)

    print("Entry", i, "of", len(en_data))

    new_entry = {'instruction': '',
                 'input': '',
                 'output': ''}

    instruction_text = en_data[i]['instruction']
    input_text = en_data[i]['input']
    output_text = en_data[i]['output']

    if check_text(instruction_text):
        new_entry['instruction'] = translate_text(instruction_text)
    else:
        new_entry['instruction'] = instruction_text

    if check_text(input_text):
        new_entry['input'] = translate_text(input_text)
    else:
        new_entry['input'] = input_text

    if check_text(output_text):
        new_entry['output'] = translate_text(output_text)
    else:
        new_entry['output'] = output_text
    
    mt_data.append(new_entry)

    # #save to file in structured way
    json.dump(new_entry, f)
    f.write("\n")


f.close()
```
